```python
# prompt: write a function that takes all the data, team and date and returns all the points score before that day

def all_points_scores_before_date(df, team, date, season = None):
  """
  Returns all the points scores for a given team before a given date.
  """

  if season is not None:
    df = df.loc[df['season'] == season]
  home_df = df.loc[(df['game_date'] < date) & ((df['home_team'] == team))]
  away_df = df.loc[(df['game_date'] < date) & ((df['away_team'] == team))]
  home_df['points'] = home_df['home_score']
  away_df['points'] = away_df['away_score']
  df = pd.concat([home_df, away_df])
  return df['points'].sum()

def all_points_scored_against_before_date(df, team, date):
  """
  Returns all the points scores for a given team before a given date.
  """
  home_df = df.loc[(df['game_date'] < date) & ((df['home_team'] == team))]
  away_df = df.loc[(df['game_date'] < date) & ((df['away_team'] == team))]
  home_df['points'] = home_df['away_score']
  away_df['points'] = away_df['home_score']
  df = pd.concat([home_df, away_df])
  return df['points'].sum()

# returns number of games team1 beat team2
def team1_beat_team2_before_date(df, team1, team2, date):
  """
  Returns the number of games team1 beat team2 before a given date.
  """
  home_wins = df.loc[(df['game_date'] < date) & (df['home_team'] == team1) & (df['away_team'] == team2) & (df['home_score'] > df['away_score'])]
  away_wins = df.loc[(df['game_date'] < date) & (df['home_team'] == team2) & (df['away_team'] == team1) & (df['away_score'] > df['home_score'])]
  return len(home_wins)

# returns total points team1 scored against team2
def team1_points_against_team2_before_date(df, team1, team2, date):
  """
  Returns the number of points team1 scored against team2 before a given date.
  """
  home_scores = df.loc[(df['game_date'] < date) & (df['home_team'] == team1) & (df['away_team'] == team2), 'home_score'].sum()
  away_scores = df.loc[(df['game_date'] < date) & (df['away_team'] == team1) & (df['home_team'] == team2), 'away_score'].sum()
  return home_scores + away_scores

all_points_scores_before_date(final_scores_df, "BAL", "2024-12-12")

def win_percentage_before_date(df, team, date):
    home_df = df.loc[(df['game_date'] < date) & (df['home_team'] == team)]
    away_df = df.loc[(df['game_date'] < date) & (df['away_team'] == team)]


    home_df['win'] = home_df['home_score'] > home_df['away_score']
    away_df['win'] = away_df['away_score'] > away_df['home_score']
    df_combined = pd.concat([home_df, away_df])

    total_games = len(df_combined)
    total_wins = df_combined['win'].sum()

    if total_games == 0:
        return 0  # Return 0 if no games played before the date

    win_percentage = total_wins / total_games
    return win_percentage

def team_recent_form(df, team, date, n=5, season=None):
    if season is not None:
        df = df.loc[df['season'] == season]

    # Filter home and away games for the given team before the specified date
    home_df = df.loc[(df['game_date'] < date) & (df['home_team'] == team)]
    away_df = df.loc[(df['game_date'] < date) & (df['away_team'] == team)]
    home_df['win'] = home_df['home_score'] > home_df['away_score']
    away_df['win'] = away_df['away_score'] > away_df['home_score']
    df_combined = pd.concat([home_df, away_df])

    # Sort by date in descending order to get the most recent games first
    df_combined = df_combined.sort_values(by='game_date', ascending=False)

    # Get the last n games (recent form)
    recent_games = df_combined.head(n)

    # Calculate the number of wins in the last n games
    total_games = len(recent_games)
    total_wins = recent_games['win'].sum()

    if total_games == 0:
        return 0  # Return 0 if no games played before the date

    recent_win_percentage = total_wins / total_games
    return recent_win_percentage

def all_points_scores_in_season(df, team, season=2024):
    """
    Returns all the points scored by a given team in the specified season.
    """

    # Filter for the specified season
    df_season = df.loc[df['season'] == season]

    # Filter home and away games for the team in the specified season
    home_df = df_season.loc[df_season['home_team'] == team]
    away_df = df_season.loc[df_season['away_team'] == team]

    # Assign points scored in home and away games
    home_df['points'] = home_df['home_score']
    away_df['points'] = away_df['away_score']

    # Concatenate home and away games
    df_combined = pd.concat([home_df, away_df])

    # Return the total points scored
    return df_combined['points'].sum()

def all_wins_in_season(df, team, season=2024):
    """
    Returns the total number of wins for a given team in the specified season.
    """

    # Filter for the specified season
    df_season = df.loc[df['season'] == season]

    # Filter home and away games for the team in the specified season
    home_df = df_season.loc[df_season['home_team'] == team]
    away_df = df_season.loc[df_season['away_team'] == team]

    # Assign wins for home and away games
    home_df['win'] = home_df['home_score'] > home_df['away_score']
    away_df['win'] = away_df['away_score'] > away_df['home_score']

    # Concatenate home and away games
    df_combined = pd.concat([home_df, away_df])

    # Return the total number of wins
    return df_combined['win'].sum()


def all_touchdowns_in_season(df, team, season=2024):
    """
    Returns the total number of touchdowns for a given team in the specified season.
    """

    # Filter for the specified season
    df_season = df.loc[df['season'] == season]

    # Filter home and away games for the team in the specified season
    home_df = df_season.loc[df_season['home_team'] == team]
    away_df = df_season.loc[df_season['away_team'] == team]

    # Assign touchdowns for home and away games
    home_df['touchdowns'] = home_df['home_score'] // 7  # Assuming 7 points per touchdown
    away_df['touchdowns'] = away_df['away_score'] // 7  # Assuming 7 points per touchdown

    # Concatenate home and away games
    df_combined = pd.concat([home_df, away_df])

    # Return the total number of touchdowns
    return df_combined['touchdowns'].sum()

def all_games_with_weather_before_date(df, team, date):
    """
    Returns the weather conditions for all games of a given team before a given date.
    """
    # Filter for games where the team is either the home or away team before the given date
    home_df = df.loc[(df['game_date'] < date) & (df['home_team'] == team)]
    away_df = df.loc[(df['game_date'] < date) & (df['away_team'] == team)]

    # Include weather condition for both home and away games
    home_df['weather'] = home_df['weather']
    away_df['weather'] = away_df['weather']

    # Concatenate the two dataframes
    df = pd.concat([home_df, away_df])

    # Return a list of the weather conditions for all games before the given date
    return df[['game_date', 'weather']].sort_values(by='game_date')


def points_per_game_in_season(df,team,season=2024):
    """
    Returns a DataFrame with each team's points per game (PPG) in the specified season.
    """

    # Filter for the specified season
    df_season = df.loc[df['season'] == season]

    # Get a list of all unique teams
    teams = pd.concat([df_season['home_team'], df_season['away_team']]).unique()

    # Initialize a dictionary to store total points and number of games for each team
    team_stats = {team: {'total_points': 0, 'games_played': 0} for team in teams}

    # Iterate over each game and update total points and games played for each team
    for _, row in df_season.iterrows():
        home_team = row['home_team']
        away_team = row['away_team']

        # Add home team's points and increment their game count
        team_stats[home_team]['total_points'] += row['home_score']
        team_stats[home_team]['games_played'] += 1

        # Add away team's points and increment their game count
        team_stats[away_team]['total_points'] += row['away_score']
        team_stats[away_team]['games_played'] += 1

    # Calculate points per game for each team
    ppg_data = {team: team_stats[team]['total_points'] / team_stats[team]['games_played']
                for team in teams}

    # Convert to a DataFrame for easier handling
    ppg_df = pd.DataFrame(list(ppg_data.items()), columns=['team', 'points_per_game'])

    # Return the DataFrame sorted by points per game
    return ppg_df.sort_values(by='points_per_game', ascending=False)

def point_differential_per_game_in_season(df, team, season=2024):
    """
    Returns a DataFrame with each team's point differential per game (PDG) in the specified season.
    """

    # Filter for the specified season
    df_season = df.loc[df['season'] == season]

    # Get a list of all unique teams
    teams = pd.concat([df_season['home_team'], df_season['away_team']]).unique()

    # Initialize a dictionary to store total point differential and number of games for each team
    team_stats = {team: {'total_point_diff': 0, 'games_played': 0} for team in teams}

    # Iterate over each game and update point differential and games played for each team
    for _, row in df_season.iterrows():
        home_team = row['home_team']
        away_team = row['away_team']
        home_score = row['home_score']
        away_score = row['away_score']

        # Calculate point differential for home and

```
